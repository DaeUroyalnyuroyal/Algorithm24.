## 1. 문제 정의

퀵 정렬

## 2. 알고리즘 설명

병합 정렬이 항목들을 위치에 따라 분할하는 것과는 달리 퀵 정렬은 항목의 값에 따라 분할 한다. 따라서 병합 정렬과 달리 리스트가 균등하게 분할 되지 않는다.

## 4. 코드 개요

함수명: quick_sort

입력:
A (리스트): 정렬할 대상 리스트
left (정수): 정렬할 부분 리스트의 시작 인덱스
right (정수): 정렬할 부분 리스트의 끝 인덱스
변환값: 없음 (주어진 리스트 A는 함수 내에서 직접 수정됩니다)

１。 만약 정렬 범위가 2개 이상의 원소를 가진 경우:

２。 partition 함수를 사용하여 리스트를 좌우로 분할하고, 중간 지점을 mid에 저장합니다.

３。 왼쪽 부분 리스트를 퀵 정렬합니다 (시작 인덱스부터 mid-1까지).

４。 오른쪽 부분 리스트를 퀵 정렬합니다 (mid+1부터 끝 인덱스까지).

## 5. 코드(py)

```python
def quick_sort(A, left, right): # A[left.right]를 오름차순으로 정렬
    if left<right:#정렬 범위가 2개 이상인 경우
        mid = partition(A, left, right)#좌우로 분할
        quick_sort(A, left, mid-1)#왼쪽 부분리스트를 퀵 정렬
        quick_sort(A, mid+1, right)#오른쪽 부분리스트를 퀵 정렬
```

## 6. 테스트

```python
def partition(A, left, right):
    low = left +1
    high = right
    pivot = A[left]
    while low <= high:
        while low <= right and A[low] <= pivot: low += 1
        while high >= left+1 and A[high] > pivot: high -= 1
        if low < high:
            A[low], A[high] = A[high], A[low]
    A[left], A[high] = A[high], A[left] # 이 부분을 while 루프 밖으로 옮깁니다.
    return high

def quick_sort(A,left,right):
    if left < right:
        mid = partition(A,left,right)
        quick_sort(A,left,mid-1)
        quick_sort(A,mid+1,right)
    
data = [5,3,8,4,9,1,6,2,7]
sorted = [0]*len(data)
print("Oringinal : ",data)
quick_sort(data,0,len(data)-1)
print("QuickSort :",data)
```
