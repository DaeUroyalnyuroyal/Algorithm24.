## 1. 문제 정의

병합알고리즘(정렬된 두 리스트의 병합)

## 2. 알고리즘 설명

병합 정렬에서는 두 개의 정렬된 부분 리스트를 병합하는 과정에서 임시 리스트가 필요하다. 병합은 각 부분 리스트의 첫 번째 항목부터 시작하며, 각 항목을 비교하여 더 작은 요소를 새로운 리스트에 복사한다. 이 과정을 반복하여 모든 항목이 복사될 때까지 진행하고, 마지막으로 다른 리스트의 남은 항목들을 모두 새로운 리스트에 복사하여 병합을 완료한다.

## 4. 코드 개요(입력 변수, 출력, 함수 설명)


함수명: merge

입력:
A (리스트): 정렬할 대상 리스트
left (정수): 병합할 부분 리스트의 시작 인덱스
mid (정수): 병합할 부분 리스트의 중간 지점 인덱스
right (정수): 병합할 부분 리스트의 끝 인덱스
변환값: 없음 (주어진 리스트 A는 함수 내에서 직접 수정됩니다)

1. 먼저, 임시 리스트의 인덱스 k를 left로 초기화하고, 왼쪽 리스트의 인덱스 i를 left로, 오른쪽 리스트의 인덱스 j를 mid + 1로 초기화합니다.
2. i가 mid 이하이고 j가 right 이하인 동안 반복하며, 두 부분 리스트를 비교합니다.
3. A[i]가 A[j] 이하이면 sorted[k]에 A[i]를 할당하고 i와 k를 1씩 증가시킵니다. 그렇지 않으면 A[j]를 할당하고 j와 k를 증가시킵니다.
4. 만약 i가 mid보다 크다면, 오른쪽 리스트에 남아 있는 레코드를 일괄 복사하고, 그렇지 않으면 왼쪽 리스트에 남아 있는 레코드를 일괄 복사합니다.
5. 마지막으로, 정렬된 부분 리스트를 원래 리스트 A의 해당 부분에 복사하여 정렬된 결과를 얻습니다.

## 5. 코드(py)

```python
def merge(A, left, mid, right):
    k = left#병합을 위한 임시 리스트의 인덱스
    i = left#왼쪽 리스트의 인덱스
    j = mid + 1#오른쪽 리스트의 인덱스
    while i<=mid and j<=right:
        if A[i] <= A[j]: #기본 연산
            sorted[k] = A[i]
            i, k = i+1, k+1
        else:
            sorted[k] = A[j]
            j, k = j+1, k+1
    if i>mid:#한쪽에 남아 있는 레코드의 일괄 복사
        sorted[k:k+right-j+1] =A[j:right+1] #리스트의 슬라이싱 이용
    else:
        sorted[k:k+mid-i+1] = A[i:mid+1] #리스트의 슬라이싱 이용
    A[left:right+1] = sorted[left:right+1] # sorted A에 다시 복사
```

병합 정렬의 분할 과정에서 n개의 원소를 2개씩 나누어 가면서 분할합니다. 이러한 분할은 logn 회 수행됩니다. 

분할된 부분 문제의 크기가 1이 되면, 각 부분 문제는 상수 시간 내에 해결됩니다. 

병합 과정에서는 각 부분 문제의 크기인 n/2만큼의 비교 연산이 수행됩니다. 따라서, 병합 정렬의 시간 복잡도는 O(nlogn)입니다.
