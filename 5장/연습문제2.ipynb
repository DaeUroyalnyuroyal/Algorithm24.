## 2. 정리 5.1을 증명해보라

1. 리스트의 합 문제의 재귀식 T(n)=2T(n/2)+1에 Master Theorem을 적용하면, 
𝑎 = 𝑏 = 2이고, 𝑓(𝑛) = 1 ∈ O(1) = 𝑂(𝑛^0)이므로 𝑑 = 0입니다. 따라서 세 번째 경우 𝑎 > 𝑏^𝑑(2 > 2^0 = 1)에 해당하며, 점근적 복잡도는 
𝑂(𝑛 log^2 𝑛) = 𝑂(𝑛)에 속합니다.

2. 만약 재귀식이 T(n)=2T(n/2)+n이라면, a=2, b=2, 𝑓(𝑛) = 𝑛 ∈ 𝑂(𝑛) = 𝑂(𝑛^1)으로 d=1입니다. 따라서 두 번째 경우 𝑎 = 𝑏^𝑑에 해당하며, 점근적 복잡도는 
   𝑂(𝑛 log⁡𝑛)입니다.

3. 마지막으로, 재귀식이 𝑇(𝑛)=3𝑇(𝑛/2)+𝑛^2라면, 𝑎=3, 𝑏=2, 𝑓(𝑛) = 𝑛^2 ∈ 𝑂(𝑛^2) = 𝑂(𝑛^2)로 𝑑=2입니다. 따라서 첫 번째 경우 𝑎 < 𝑏^𝑑에 해당하며, 점근적 복잡도는 𝑂(𝑛^2)가 됩니다.
마스터 정리는 많은 순환 관계식의 점근적 복잡도를 계산하는 데 유용하게 사용될 수 있지만, 모든 순환식에 적용할 수 있는 것은 아닙니다.
