## 1. 문제 정의


리스트에 n개의 항목이 들어 있다. 이 리스트에서 “탐색기”를 가진 항목을 찾아라. 만약 찾는 항목이 있으면 그 항목의 인덱스를 반환하고, 없으면 -1을 반환하라. 단, 리스트의 항목들은 정렬되어 있지 않다.

## 2. 알고리즘 설명

입력리스트의 척 항목부터 순서대로 하나씩 탐색키와 비교한다.
항목 하나씩 검사하다가 성공하면 인덱스를 반환, 살패하면(찾는 항목이 없으면) -1을 반환한다.


## 3. 손으로 푼 예제

## 4. 알고리즘 코드 개요

1. 함수 seq_search는 리스트 A에서 key를 순차적으로 찾아 해당하는 인덱스를 반환합니다.

2. 리스트 A의 길이만큼 반복합니다.

3. 현재 요소가 key와 같은지 확인합니다.

4. 같다면 해당하는 인덱스를 반환합니다.

5. key가 리스트 A 안에 존재하지 않는 경우 -1을 반환합니다.

## 5. 알고리즘 코드(py)

```python
def seq_search(A, key):  #리스트 A에서 key를 찾아 해당 인덱스를 반환
    for i in range(len(A)):  # 리스트 A를 순회하며
        if A[i] == key:  # 요소가 key와 같으면
            return i  # 해당하는 인덱스를 반환
    return -1  # key가 리스트 A 안에 없으면 -1 반환
```

## 6. 테스트 코드(py)

```python
def seq_search(A, key):  #리스트 A에서 key를 찾아 해당 인덱스를 반환
    for i in range(len(A)):  # 리스트 A를 순회하며
        if A[i] == key:  # 요소가 key와 같으면
            return i  # 해당하는 인덱스를 반환
    return -1  # key가 리스트 A 안에 없으면 -1 반환

print("인덱스",seq_search([1, 3, 4, 6, 7, 9, 8, ], 4), "에 있다.")
# 리스트 [1, 3, 4, 6, 7, 9, 8]에서 4를 찾아 해당하는 인덱스를 출력합니다.
```
