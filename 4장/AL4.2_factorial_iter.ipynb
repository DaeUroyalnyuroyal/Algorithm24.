## 1. 문제 정의


팩토리얼(상향식 축소 정복 기법)

## 2. 알고리즘 설명

"축소 정복" 기법 중 하나인 "상향식 접근"을 사용하여 주어진 양의 정수 `n`의 팩토리얼을 계산하는 알고리즘입니다. 이 알고리즘은 반복적인 구조를 이용하여 문제를 해결합니다.


## 3. 손으로 푼 예제

## 4. 코드 개요(입력 변수, 출력, 함수 설명)


함수명: factorial_iter

반복문을 사용하여 `n!`을 계산하는 "반복적 접근" 알고리즘입니다. 반복문을 통해 1부터 `n`까지의 모든 정수를 곱해나가는 방식으로 구현되어 있습니다.

입력: n (양의 정수)

변환값: 1 또는 result

## 5. 코드 (PY)

```python
def factorial_iter(n) :#축소 정복 기법: 상향식 접근
    result = 1
    for k in range (1,n+1) : #반복구조 이용
        result = result * k
    return result
```

## 6. 테스트 코드(PY)

```python
def factorial_iter(n) :#축소 정복 기법: 상향식 접근
    result = 1
    for k in range (1,n+1) : #반복구조 이용
        result = result * k
    return result

print(factorial_iter(5))
```

## 8. 복잡도 분석

O(n)입니다. 왜냐하면 반복문은 n번 실행되기 때문입니다. 각 반복에서의 연산은 상수 시간이 소요되므로, 전체 시간 복잡도는 입력된 값 n에 비례합니다.
