## 1. 문제 정의


n개의 항목을 가진 리스트를 오름 차순으로 정렬하라.

## 2. 알고리즘 설명

A[0~i-1]에 넣는 경우를 생각해 볼때 먼저 맨 왼쪽항목부터 A[i]와 하나씩 비교하여 오른쪽으로 방향으로 진행하는데 현재 항목이 A[i]보다 작으면 계속 진행하고 크거나 같으면 멈춘다 마지막으로 위치에A[i]를 끼워넣으면 된다.

(정렬해야할 항목이 하나가 될때까지 반복되는데, 항목이 하나가 되면 이미 정렬된 상태가 된다.)

## 4. 코드 개요(입력 변수, 출력, 함수 설명)


함수명: insertion_sort 

주어진 리스트를 정렬하는 데 사용되는 함수입니다. 이 함수는 삽입 정렬(Insertion Sort) 알고리즘을 사용하여 리스트를 오름차순으로 정렬합니다. 

입력: A (정렬되지 않은 형태의 리스트를 나타내는 변수 이름)

변환값: 없음(print로 출력)

1. 정렬할 리스트 A를 입력한다.

2. A[i - 1]부터 왼쪽으로 비교를 진행하는데, key보다 크면 계속 반복한다.

3. 항복이 key보다 크면 오른쪽으로 한칸 이동한다.

4. 맨 마지막에 key(A[i])를 제 위치에 복사한다.

5. 결과 출력을 위한 함수를 호출한다.

## 5. 코드 (PY)

```python
def insertion_sort(A) : 
    n = len(A) # 입력의 크기 
    for i in range(1, n) : # 외부 루프: 1. 2. ··· n-1 
        key = A[i] # A[i]를 key에 저장 
        j = i-1 # 항목 A[i-1]부터 검사 
        while j>=0 and A[j] > key : # 내부 루프 
            A[j + 1] = A[j] # 항목들을 오른쪽으로 한 칸씩 이동 
            j = j - 1 # 위치를 왼쪽으로 이동 
        A[j + 1] = key # 항목 A[i]를 제 위치에 삽입 
        printStep(A, i)
```

## 8. 복잡도 분석

입력 배열을 정렬하기 위해 삽입 정렬 알고리즘은 외부 루프와 내부 루프를 모두 사용합니다. 외부 루프는 입력 배열의 첫 번째 요소부터 마지막 요소까지 n번 실행됩니다. 내부 루프는 외부 루프에서 현재 요소의 앞쪽 요소들 중에서 큰 값을 만나면 계속해서 반복됩니다.

이때, 외부 루프와 내부 루프가 모두 n번 실행되는 이유는 입력 배열의 각 요소에 대해 삽입 정렬을 수행하기 때문입니다. 외부 루프는 현재 요소를 선택하고, 내부 루프는 현재 요소를 올바른 위치에 삽입하기 위해 앞쪽 요소들과 비교하며 반복합니다.

따라서, 삽입 정렬 알고리즘은 최악의 경우 모든 요소에 대해 n번씩 내부 루프를 실행하게 되므로, 전체적인 실행 시간은 입력 배열의 길이 n에 대해 제곱인 O(n^2)입니다.
